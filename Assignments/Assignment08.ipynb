```python
import matplotlib.pyplot as plt
import math 
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
```

```python
data = [
        'TB3MS', # us interest rate
        'CLVMNACSCAB1GQSE',# sweden GDP
        'SWECPIALLMINMEI', # swedish CPI
        'IR3TIB01SEM156N', # interest rate sweden 
        'CCRETT01SEM661N'# real effective interest rate 
        ]
```

```python
df = pdr.DataReader(
    data,
    "fred",
    start="1992-01-01",
    end="2025-07-01",
).astype(float)
```

```python
df['CLVMNACSCAB1GQSE'] = df['CLVMNACSCAB1GQSE'].interpolate(method='linear')
df['SWECPIALLMINMEI'] = df['SWECPIALLMINMEI'].interpolate(method='linear')
```

```python
df = df.resample('Q').mean()
```

```python
df['pit'] = 400 * np.log(df['SWECPIALLMINMEI']/ df['SWECPIALLMINMEI'].shift(1))
```

```python
df = df.dropna()
```

```python
df = df.rename(columns = {
    'TB3MS': 'i_star',
    'CLVMNACSCAB1GQSE' : 'yt',
    # 'SWECPIALLMINMEI': 'pit',
    'IR3TIB01SEM156N': 'it',
    'CCRETT01SEM661N': 'et'

})
```

```python
df['yt'] = np.log(df['yt'])
df['et'] = np.log(df['et'])
```

```python
df['d1t'] = (df.index == '1993-03-31').astype(int)
df['d2t'] = (df.index == '1995-12-31').astype(int)
df['d3t'] = (df.index == '1992-09-30').astype(int)
```
